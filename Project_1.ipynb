```python
data[data.imdb_id.str.match('tt', na=False)] #удостоверился, что все 1889 позиций imdb_id начинаются с 'tt'
```

```python
data.imdb_id = data.imdb_id.apply(lambda x:int(x[2:])) #убрал из imdb_id начало 'tt', перевел в значение int64
```

```python
def change_rd(x): #формируем функцию для приведения к удобному формату даты релиза
    d = x.split('/')
    y = d[1]+'.'+d[0]+'.'+d[2]
    return y
data.release_date = data.release_date.apply(change_rd)
```

```python
def clean_object(x): #формируем функцию по отчистке текстовых данных от знаков и литеров не из английского алфавита, а так же оставляем пробелы и знаки препинания
    y = ''.join(z for z in x if z in string.ascii_letters or z in '''|,.&;:?!-`()''' or z in ' ')
    return y
data.original_title = data.original_title.apply(clean_object)
data.cast = data.cast.apply(clean_object)
data.director = data.director.apply(clean_object)
data.tagline = data.tagline.apply(clean_object)
data.overview = data.overview.apply(clean_object)
data.genres = data.genres.apply(clean_object)
data.production_companies = data.production_companies.apply(clean_object)
```

```python
data['profit'] = data.revenue - data.budget #добавил колонку profit
columns = data.columns.values; print(columns) #переставил колонки в удобном порядке
data = data[['imdb_id', 'original_title', 'tagline', 'overview', 'cast', 'director', 'genres', 'production_companies', 
             'budget', 'revenue', 'profit', 'runtime', 'release_date', 'release_year', 'vote_average']]
data.head()
```

```python
def quantity_symb(x): #функция подсчета количества символов
    quantity = 0
    for i in x:
        quantity += 1
    return quantity
```

```python
def quantity_word(x): #функция подсчета кол-ва слов
    return len(x.split(' '))
```

```python
list_mounths = [] #список для всех месяцев всех фильмов
for i in data.release_date: #вырезаем месяц из каждого фильма, добавляем в список
    mounth = i.split('.')[1]
    list_mounths.append(int(mounth))
data['release_mounth'] = list_mounths #присваеваем новой колонке месяцев релиза фильмов значения списка
```

# 11. Какого жанра фильмов больше всего?

```python
all_genres = []
for i in data.genres: #создал единые список всех жанров всех фильмов
    y = i.split('|')
    all_genres = all_genres + y

c = collections.Counter() #создал счетчик видов жанров
for genre in all_genres:
    c[genre] += 1
print(c)
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
dict_actors = {} #формируем словарь, циклическую проверку пар актеров в одном фильме
for i in data.cast:
    a = i.split('|')
    counter = 0
    for ii in range(len(a)-2):
        for iii in range(1, len(a)-1):
            b = a[ii] + ' & ' + a[iii + counter]
            if b in dict_actors:
                dict_actors[b] += 1
            else:
                dict_actors[b] = 1
            if iii + counter == len(a)-1:
                break
        counter += 1
                                        #пример работы цикла для 3 актеров в фильме (по индексам в списке а):
                                        #первый цикл от и0 до и2 (и - индекс):
                                            #второй цикл от 1 до 3:
                                                #проверка актеров и0 + и(1+каунтер0), и0+и(2+к0), и0+и(3+к0)
                                        #и т.д.

list_dict_actors = list(dict_actors.items()) #сортировка по ключу
list_dict_actors.sort(key=lambda i: i[1])

print(list_dict_actors)
```
